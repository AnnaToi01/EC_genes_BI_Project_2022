# Phylogenetic analysis

For phylogenetic analysis we use the file with amino acid FASTA sequences generated in the folder `OrthoGroups_analysis`.

Versions:
* MUSCLE - v5.1
* MAFFT - v7.505
* ClustalO - 1.2.3
* Ugene - v41
* [IQ-Tree v2.0.3](https://anaconda.org/bioconda/iqtree)

Coverage is universally bad, but we decided to use ClustalO to build the tree.

## MUSCLE

```python
!muscle -align conc_protein_seq.fa -output alig_muscle.fa
```

# ClustalO

**1. ClustalO Alignment**

```python
!clustalo --infile=conc_protein_seq.fa --outfile=alig_clustalo.fa --verbose
```

```python
!head alig_clustalo.fa
```

Alignment coverage as visualized in UGENE.

![alig_clustalo.png](alig_clustalo.png)

**2. Using the ClustalO alignment to build a tree**

#### IQTREE


```python
!iqtree -m MFP -s alig_clustalo.fa --prefix clustalo_not_trimmed_iqtree_bootstrap -o ERN08734_Amborella_trichopoda -B 1000 -alrt 1000 -abayes
```

## Trimming

We also tried trimming the alignment with trimAl v1.4.rev15. We tried `gappyout` and `strict` trimming (`strict-plus` generated same result as `strict`).

**Gappyout**

```python
!trimal -in alig_clustalo.fa -out gappyout_trimmed_alig_clustalo.fa -gappyout
```

```python
!iqtree -m MFP -s gappyout_trimmed_alig_clustalo.fa --prefix clustalo_gappyout_trimmed_iqtree_bootstrap -o ERN08734_Amborella_trichopoda -B 1000 -alrt 1000 -abayes
```

Coverage:
